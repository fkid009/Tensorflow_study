# Package

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.metrics import Mean, SparseTopKCategoricalAccuracy
```

# Data Generate

```python
train_ds, ds_info = tfds.load(name = "mnist",
                              shuffle_files= True,
                              as_supervised= True,
                              split = "train",
                              with_info= True)
```

```python
def normalization(images, labels):
    images = tf.cast(images, tf.float32) / 255.
    return [images, labels]
```

```python
train_batch_size = 10
train_ds = train_ds.map(normalization).batch(train_batch_size)
```

# Model

```python
n_layer = 10
units = [10] * n_layer
```

```python
model = Sequential()
model.add(Flatten())
for layer_idx in range(n_layer - 1):
    model.add(Dense(units = units[layer_idx], activation = "sigmoid"))
model.add(Dense(units = 10, activation = "softmax"))
```

```python
model.build(input_shape= (None, 28, 28, 1))
model.summary()
```

# Initiation

```python
loss_object = SparseCategoricalCrossentropy()
optimizer = SGD()
```

# Train (1 Epoch)

```python
train_ds = iter(train_ds)
images, labels = next(train_ds)
```

```python
with tf.GradientTape() as tape:
    predictions = model(images)
    loss = loss_object(labels, predictions)
gradients = tape.gradient(loss, model.trainable_variables)
```

# Visualization of Gradient Vanishing

## overall gradient

```python
cmap = cm.get_cmap("rainbow", lut = n_layer + 1)

fig, ax = plt.subplots(figsize = (15, 10))
ax.set_yscale("log")
for grad_idx, grad in enumerate(gradients[::2]):
    if grad_idx >= 1:
        grad_abs = np.abs(grad.numpy().flat)
        x_range = np.arange(len(grad_abs))
        ax.plot(grad_abs,
                label = f"layer {grad_idx}",
                color = cmap(grad_idx),
                alpha = 0.9)

ax.legend(bbox_to_anchor = (1, 0.5),
          loc = "center left",
          fontsize = 20,
          ncol = 2)
fig.tight_layout()
```

## average gradient

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.set_yscale("log")
grad_means = []
for grad_idx, grad in enumerate(gradients[::2]):
    if grad_idx >= 1:
        grad_abs = np.abs(grad.numpy().flat)
        grad_means.append(np.mean(grad_abs))

ax.plot(grad_means)
fig.tight_layout()
```
